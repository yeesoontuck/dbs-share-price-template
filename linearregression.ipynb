# Linear regression model using DBS csv data

- import CSV
- X = SGD price column
- y = DBS share price column
- fit the linear regression model
- get the coefficient
- get the y-intercept

```python
import pandas as pd
```

```python
df = pd.read_csv('data/DBS_SingDollar.csv')
df
```

```python
X = df.loc[:,["SGD"]]
# not X = df['SGD'], because sklearn needs a 2D DataFrame as input
X
```

```python
y = df.loc[:, ["DBS"]]
y
```

```python
from sklearn import linear_model

reg = linear_model.LinearRegression()
reg.fit(X, y)
```

```python
coef = reg.coef_
# array([[-50.60094302]])
```

```python
coef = coef[0][0]
```

```python
intercept = reg.intercept_
# array([90.22858515])
```

```python
intercept = intercept[0]
```

```python
y_pred = reg.predict(X)
```

```python
from sklearn.metrics import mean_squared_error
print("Root Mean squared error: %.2f" % mean_squared_error(y, y_pred)**0.5)
```

```python
import matplotlib.pyplot as plt

plt.scatter(X, y, color='blue')            # actual data
plt.plot(X, y_pred, color='red')           # regression line
plt.xlabel("SGD")
plt.ylabel("DBS")
plt.title("Linear Regression: SGD vs DBS")
plt.show()
```

```python
print(f"Coefficient: {coef:.1f}")
print(f"y-intercept: {intercept:.1f}")
```
